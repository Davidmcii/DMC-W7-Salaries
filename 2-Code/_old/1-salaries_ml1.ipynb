```python
def funct(x):

    if 'Developer' in x:
        function='Developer'      
    if 'Engineer' in x:
        function='Engineer'
    if 'Analyst' in x:
        function='Analyst'	
    if 'Architect' in x:
        function='Architect'
    if 'Scientist' in x:
        function='Scientist'
    if 'Researcher' in x:
        function='Researcher'
    if 'Consultant' in x:
        function='Consultant'
    if 'Staff' in x:
        function='Staff'
    if 'Principal' in x:
        function='Principal'
    if 'Lead' in x:
        function='Lead'
    if 'Manager' in x:
        function='Manager'
    if 'Specialist' in x:
        function='Specialist'
    if 'Director' in x:
        function='Director'
    if 'Head' in x:
        function='Head'
    if 'Business' in x:
        function='Business'
    if 'Fina' in x:
        function='Finance'
    
    return function    

tog.job_title=tog.job_title.apply(lambda x: funct(x))

tog.job_title.unique()

```

```python
from sklearn.linear_model import LogisticRegression 

from sklearn.model_selection import train_test_split as tts 
```

```python
X_train, X_test, y_train, y_test=tts(X, y, random_state=42, test_size=0.20)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression as LinReg
from sklearn.linear_model import Lasso        # regularizacion L1
from sklearn.linear_model import Ridge        # regularizacion L2
from sklearn.linear_model import ElasticNet   # regularizacion L1+L2


# se inician los modelos
linreg=LinReg()
lasso=Lasso()
ridge=Ridge()
elastic=ElasticNet()
```

```python
from sklearn.svm import SVR  # support vector regressor

svr=SVR()

from sklearn.ensemble import RandomForestRegressor as RFR  
from sklearn.tree import ExtraTreeRegressor as ETR

rfr=RFR()
etr=ETR()

from sklearn.ensemble import GradientBoostingRegressor as GBR

from xgboost import XGBRegressor as XGBR

from catboost import CatBoostRegressor as CTR

from lightgbm import LGBMRegressor as LGBMR


gbr=GBR()
xgbr=XGBR()
ctr=CTR()
lgbmr=LGBMR()
```

```python
linreg.fit(X_train, y_train)
lasso.fit(X_train, y_train)
ridge.fit(X_train, y_train)
elastic.fit(X_train, y_train)
svr.fit(X_train, y_train)
rfr.fit(X_train, y_train)
etr.fit(X_train, y_train)
gbr.fit(X_train, y_train)
xgbr.fit(X_train, y_train)
ctr.fit(X_train, y_train, verbose=0)
lgbmr.fit(X_train, y_train)
```
