```python
X_train, X_test, y_train, y_test=tts(X, y, random_state=42, test_size=0.3)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
modelos=[linreg, lasso, ridge, elastic, svr, rfr, etr, gbr, xgbr, ctr,lgbmr]

modelos_name=['linreg', 'lasso', 'ridge', 'elastic', 'svr', 'rfr', 'etr', 'gbr', 'xgbr', 'ctr', 'lgbmr']

mse_testeo=[]
mse_train=[]
erre_testeo=[]
erre_train=[]

for i,m in enumerate(modelos):
    if modelos_name[i]=='ctr':
        m.fit(X_train, y_train, verbose=0)

    else:
        m.fit(X_train, y_train)
    
    m.predict(X_test)[:10]
    
    y_pred=m.predict(X_test)
    y_pred_train=m.predict(X_train)
    
    
    mse_testeo.append(mse(y_test, y_pred, squared=False))
    mse_train.append(mse(y_train, y_pred_train, squared=False))
    erre_testeo.append(r2(y_test, y_pred))
    erre_train.append(r2(y_train, y_pred_train))
    

analisis= pd.DataFrame({"modelos" : modelos_name, "mse_testeo" : mse_testeo, "mse_train" : mse_train, "erre2_testeo" : erre_testeo, "erre2_train" : erre_train})

analisis

#resultado= dict(zip(modelos_name, list(zip(mse_testeo,mse_train,erre_testeo,erre_train))))

#resultado
```
